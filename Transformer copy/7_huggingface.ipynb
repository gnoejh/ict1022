```python
config = {
    # "task": "translation",  
    # "task": "summarization", 
     "task": "text-generation", 
    # "task": "sentiment-analysis", 
    # "task": "question-answering", 
    # "task": "ner", 
    # "task": "text-classification", 
    # "task": "language-modeling", 
    # "task": "paraphrasing", 
    # "task": "speech-recognition", 
    # "task": "machine-reading-comprehension"
}

tasks = {
    "summarization": {
        "model": "facebook/bart-large-cnn",
        "tokenizer": "facebook/bart-large-cnn"
    },
    "text-generation": {
        "model": "gpt2",
        "tokenizer": "gpt2"
    },
    "sentiment-analysis": {
        "model": "distilbert-base-uncased-finetuned-sst-2-english",
        "tokenizer": "distilbert-base-uncased-finetuned-sst-2-english"
    },
    "question-answering": {
        "model": "distilbert-base-cased-distilled-squad",
        "tokenizer": "distilbert-base-cased-distilled-squad"
    },
    "ner": {
        "model": "dbmdz/bert-large-cased-finetuned-conll03-english",
        "tokenizer": "dbmdz/bert-large-cased-finetuned-conll03-english"
    },
    "text-classification": {
        "model": "bert-base-uncased",
        "tokenizer": "bert-base-uncased"
    },
    "language-modeling": {
        "model": "gpt2",
        "tokenizer": "gpt2"
    },
    "paraphrasing": {
        "model": "tuner007/pegasus_paraphrase",
        "tokenizer": "tuner007/pegasus_paraphrase"
    },
    "speech-recognition": {
        "model": "facebook/wav2vec2-base-960h",
        "tokenizer": "facebook/wav2vec2-base-960h"
    },
    "machine-reading-comprehension": {
        "model": "deepset/roberta-base-squad2",
        "tokenizer": "deepset/roberta-base-squad2"
    },
    "translation": {
        "model": "Helsinki-NLP/opus-mt-en-de",
        "tokenizer": "Helsinki-NLP/opus-mt-en-de"
    }
}

from transformers import pipeline
import torch

# Initialize the model and tokenizer based on the task
task = config["task"]
model_name = tasks[task]["model"]
tokenizer_name = tasks[task]["tokenizer"]

device = 0 if torch.cuda.is_available() else -1
nlp_pipeline = pipeline(task, model=model_name, tokenizer=tokenizer_name, device=device)

# Example usage
if task == "summarization":
    text = (
        "Artificial intelligence (AI) is intelligence demonstrated by machines, "
        "in contrast to the natural intelligence displayed by humans and animals. "
        "Leading AI textbooks define the field as the study of 'intelligent agents': "
        "any device that perceives its environment and takes actions that maximize "
        "its chance of successfully achieving its goals."
    )
    summary = nlp_pipeline(text, max_length=50, min_length=10)
    print(summary)
elif task == "text-generation":
    prompt = "Once upon a time"
    generated_text = nlp_pipeline(prompt, max_length=50, pad_token_id=50256, truncation=True)  # 50256 is the pad_token_id for GPT-2
    print(generated_text)
elif task == "sentiment-analysis":
    text = "I love using HuggingFace Transformers!"
    sentiment = nlp_pipeline(text)
    print(sentiment)
elif task == "question-answering":
    context = "HuggingFace is creating a tool that democratizes AI."
    question = "What is HuggingFace creating?"
    answer = nlp_pipeline(question=question, context=context)
    print(answer)
elif task == "ner":
    text = "HuggingFace is based in New York City."
    entities = nlp_pipeline(text)
    print(entities)
elif task == "text-classification":
    text = "HuggingFace is amazing!"
    classification = nlp_pipeline(text)
    print(classification)
elif task == "language-modeling":
    prompt = "In a distant future"
    generated_text = nlp_pipeline(prompt, max_length=50)
    print(generated_text)
elif task == "paraphrasing":
    text = "Can you rephrase this sentence?"
    paraphrase = nlp_pipeline(text)
    print(paraphrase)
elif task == "speech-recognition":
    audio_file = "path_to_audio_file.wav"
    transcription = nlp_pipeline(audio_file)
    print(transcription)
elif task == "machine-reading-comprehension":
    context = "HuggingFace is creating a tool that democratizes AI."
    question = "What is HuggingFace creating?"
    answer = nlp_pipeline(question=question, context=context)
    print(answer)
elif task == "translation":
    text = "Hello, how are you, Hong?"
    translation = nlp_pipeline(text)
    print(translation)
```
