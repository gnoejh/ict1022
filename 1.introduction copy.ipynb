# Chapter 1: Introduction to Deep Learning

## **1.1 What is Deep Learning?**

### **Definition and Core Idea**
Deep learning is a subset of machine learning that utilizes neural networks with multiple layers. It enables the modeling of complex patterns in data by leveraging large datasets and significant computational power.

```mermaid
graph TD
A[Input Layer] --> B[Hidden Layer 1]
B --> C[Hidden Layer 2]
C --> D[Output Layer]
```

**Key Points:**
- Models data hierarchically.
- Requires large datasets and computational resources.

### **1.1.1 Neural Networks and Layers**
Neural networks consist of interconnected layers of nodes (neurons) that process input data through weighted connections and activation functions.

```math
z^{(l)} = W^{(l)}a^{(l-1)} + b^{(l)}
a^{(l)} = \sigma(z^{(l)})
```

**Code Example:**
```python
import torch
import torch.nn as nn

# Define a simple feedforward neural network
class SimpleNN(nn.Module):
    def __init__(self):
        super(SimpleNN, self).__init__()
        self.fc1 = nn.Linear(10, 50)  # Input to hidden layer
        self.fc2 = nn.Linear(50, 1)  # Hidden to output layer

    def forward(self, x):
        x = torch.relu(self.fc1(x))
        x = torch.sigmoid(self.fc2(x))
        return x

# Instantiate the model
model = SimpleNN()
print(model)
```

## **1.4 Applications of Deep Learning**

### **1.4.1 Image and Video Analysis**

```mermaid
graph TB
A[Image Input] --> B[Convolutional Neural Network]
B --> C[Object Detection]
B --> D[Segmentation]
```

#### Example Code: Object Detection
```python
from torchvision import models

# Load pre-trained Faster R-CNN model
model = models.detection.fasterrcnn_resnet50_fpn(pretrained=True)
model.eval()

# Process image
from PIL import Image
from torchvision.transforms import functional as F
image = Image.open("example.jpg")
input_tensor = F.to_tensor(image).unsqueeze(0)

# Run model
predictions = model(input_tensor)
print(predictions)
```

### **1.4.2 Natural Language Processing**

#### Transformer for Machine Translation
```python
from transformers import MarianMTModel, MarianTokenizer

tokenizer = MarianTokenizer.from_pretrained("Helsinki-NLP/opus-mt-en-fr")
model = MarianMTModel.from_pretrained("Helsinki-NLP/opus-mt-en-fr")

text = "Deep learning transforms AI."
tokens = tokenizer(text, return_tensors="pt")
translation = model.generate(**tokens)
print(tokenizer.decode(translation[0], skip_special_tokens=True))
```
